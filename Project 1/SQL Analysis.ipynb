## Establish a database Connection and SQL Queries

```python
#Create if not yet or connect a database in sqlite3
con =sqlite3.connect("Data_science.db")
```

```python
#create a cursor object
curso=con.cursor()
```

##### Save the data to sqlite3 database

```python
# After rertrieving the dataset from csv to pandas dataframe and creating the data_science database save it can now be save to sqlite3 database.
# The the pandas dataframe to sql database
df.to_sql("data_scientist",con, index = False, method = 'multi')
```

##### Table 1 Top 10 highest paid data science role in U.S

```python
%sql SELECT job_title,salary_in_usd,company_location FROM data_scientist WHERE company_location = 'US' ORDER BY salary_in_usd DESC LIMIT 10;
```

##### NUMBER OF DATA SCIENTIST  ROLE

```python
%sql SELECT COUNT(job_title) FROM data_scientist WHERE job_title = 'Data Scientist';
```

##### AVERAGE SALARY OF DATA SCIENTIST ROLE IRRESPECTIVE OF SPECIFIC JOB TASK IN COMPANY OF SMALL SIZE


```python
%sql SELECT AVG(salary_in_usd) FROM data_scientist WHERE company_size = 'S';
```

###### AVERAGE SALARY OF DATA SCIENTIST ROLE IRRESPECTIVE OF SPECIFIC JOB TASK IN COMPANY OF MEDIUM SIZE

```python
%sql SELECT AVG(salary_in_usd) FROM data_scientist WHERE company_size = 'M';
```

##### AVERAGE SALARY OF DATA SCIENTIST ROLE IRRESPECTIVE OF SPECIFIC JOB TASK IN COMPANY OF LARGE SIZE

```python
%sql SELECT AVG(salary_in_usd) FROM data_scientist WHERE company_size = 'L';
```
